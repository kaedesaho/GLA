```python
dataset = load_dataset("instdin/institutional-books-1.0-metadata", split="train")
```

```python
threshold_counts = {50: 0, 60: 0, 70: 0, 80: 0, 90: 0}

for i, row in enumerate(dataset):
    if row.get("language_gen") != "gla":
        continue
    
    dist = row.get("language_distribution_gen", {})
    languages = dist.get("language", [])
    proportions = dist.get("proportion", [])

    if "gla" in languages:
        gla_index = languages.index("gla")
        gla_proportion = proportions[gla_index]

        for threshold in threshold_counts:
            if gla_proportion > threshold:
                threshold_counts[threshold] += 1

for threshold, count in sorted(threshold_counts.items()):
    print(f"> {threshold}% gla: {count} books")
```
